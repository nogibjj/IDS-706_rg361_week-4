# Descriptive Statistics Using Polars - Jupyter Notebook

```python
# Importing required libraries
import polars as pl
import matplotlib.pyplot as plt
from lib import select_col, summary_stats
```

```python
# Loading the data into a dataframe
fname = "../../resources/blood_pressure.csv"
df = pl.read_csv(fname)
```

```python
# view the first 5 rows of the dataframe
print(df.head(5))
```

```python
# use the select_col function from lib to select the columns of interest
col_name = select_col(
    df
)  # it will default to the last column since we are not specifying any column name

# check if error code(-1) is received, if no error code-proceed further
if col_name == -1:
    exit()
```

```python
# use the summary_stats function from lib to get summ. stats
result = summary_stats(df, col_name)
```

```python
# create Histogram and display it
plt.hist(df.select(col_name))
plt.ylabel("Count of " + col_name)
plt.xlabel(col_name)
plt.title("Data Loaded from : " + fname + "\n")
plt.show()
```

```python
# print the values of the summary statistics
print("The mean of {} column is : {}".format(col_name, result[0]))
print("The median of {} column is : {}".format(col_name, result[1]))
print("The standard deviation of {} column is : {}".format(col_name, result[2]))
```
